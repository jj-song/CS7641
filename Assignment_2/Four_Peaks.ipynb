#### Utility Functions

```python
def get_first_iteration(curve, fitness):
    result=np.where(curve==fitness)
    first = np.amin(result[0])
    return first+1
```

#### Problem Declaration

```python
'''
Initialize Objects: Four Peaks
    - define fitness function
    - define define an optimization problem
    
Sources:
- https://mlrose.readthedocs.io/en/stable/source/fitness.html#fitness
    
Problem Description:
    example problem: [1,1,1,0,1,0,0,1,0,0,0,0]
            threshold percent: 0.2
    - fitness is calculated by adding: (you can only add if number of 0's and 1's are both greater than threshold number)
        a = the number of leading 0's or trailing 1's (whichever one is larger) => 4
        b = length of problem space times t_pct => 12 * 0.2 = 2.4
        total = 4 + 12 = 16

'''

fitness = mlrose.FourPeaks(t_pct=0.10)

# easy
# problem = mlrose.DiscreteOpt(length = 10, fitness_fn = fitness, maximize = True, max_val=2)

# medium
# problem = mlrose.DiscreteOpt(length = 75, fitness_fn = fitness, maximize = True, max_val=2)

# hard
# problem = mlrose.DiscreteOpt(length = 100, fitness_fn = fitness, maximize = True, max_val=2)
```

#### Randomized Hill Climbing

```python
'''
Randomized Hill Climbing

Sources:
- https://mlrose.readthedocs.io/en/stable/source/algorithms.html#mlrose.algorithms.random_hill_climb

'''

# initialize problem
problem = mlrose.DiscreteOpt(length = 75, fitness_fn = fitness, maximize = True, max_val=2)

# start timer for testing
start = time.time()
# select and run randomized optimization algorithm
best_state, best_fitness, fitness_curve = mlrose.random_hill_climb(problem, max_attempts=1000, max_iters=5000, restarts=1000,
                                                     random_state=37, curve=True)
# end timer for testing
end = time.time()
# print total time for testing
print(f"Total Testing Time: {round(end-start, 5)}s")

print(best_state)
print(best_fitness)

# plot and show
plt.plot(fitness_curve[:,0], color="red", label="Randomized Hill Climbing")
plt.title("Fitness Score vs. Iterations")
plt.xlabel("Iterations"), plt.ylabel("Fitness"), plt.legend(loc="best")
plt.tight_layout()
plt.grid()
plt.show()
```

```python
'''
Comparison of All 4 Search Algorithms

'''
# initialize problem
problem = mlrose.DiscreteOpt(length = 80, fitness_fn = fitness, maximize = True, max_val=2)

# random hill climb
# start timer for testing
start = time.time()
fc_rhc = mlrose.random_hill_climb(problem, max_attempts=1000, max_iters=5000, restarts=1000,
                                                     random_state=37, curve=True)[2]
# end timer for testing
end = time.time()
# print total time for testing
print(f"RHC - Total Testing Time: {round(end-start, 5)}s")


# simulated annealing
# start timer for testing
start = time.time()
# define decay schedule (ArithDecay, GeomDecay, ExpDecay)
schedule = mlrose.ExpDecay(init_temp=300, exp_const=0.005)
fc_sa = mlrose.simulated_annealing(problem, schedule=schedule, max_attempts=1000, max_iters=5000,
                                                     random_state=37, curve=True)[2]
# end timer for testing
end = time.time()
# print total time for testing
print(f"SA - Total Testing Time: {round(end-start, 5)}s")


# genetic algorithm
# start timer for testing
start = time.time()
fc_ga = mlrose.genetic_alg(problem, pop_size=500, mutation_prob=0.6, max_attempts=1000, max_iters=5000,
                                                    random_state=37, curve=True)[2]
# end timer for testing
end = time.time()
# print total time for testing
print(f"GA - Total Testing Time: {round(end-start, 5)}s")


# MIMIC
# start timer for testing
start = time.time()
fc_mimic = mlrose.mimic(problem, pop_size=4000, keep_pct=0.30, max_attempts=1000, max_iters=5000,
                                                    random_state=37, curve=True)[2]
# end timer for testing
end = time.time()
# print total time for testing
print(f"MIMIC - Total Testing Time: {round(end-start, 5)}s")


# plot and show
plt.plot(fc_rhc[:,0], color="r", label="Random Hill Climb")
plt.plot(fc_sa[:,0], color="b", label="Simulated Annealing")
plt.plot(fc_ga[:,0], color="g", label="Genetic Algorithm")
plt.plot(fc_mimic[:,0], color="m", label="MIMIC")
plt.title("Four Peaks - Fitness Score vs. Iterations")
plt.xlabel("Iterations"), plt.ylabel("Fitness"), plt.legend(loc="best")
plt.tight_layout()
plt.grid()
plt.show()
```
